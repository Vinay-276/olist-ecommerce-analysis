```python
orders = pd.read_csv(
    r"C:\Users\Vinay\Downloads\olist_orders_dataset.csv.zip",
    compression='zip'
)
print(orders.head())
```

```python
# Convert dates
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])

# Remove missing delivery rows
orders = orders[orders['order_delivered_customer_date'].notna()]

# Create delivery_days
orders['delivery_days'] = (
    orders['order_delivered_customer_date'] - 
    orders['order_purchase_timestamp']
).dt.days

# Create delay
orders['delivery_delay_days'] = (
    orders['order_delivered_customer_date'] - 
    orders['order_estimated_delivery_date']
).dt.days

# Create is_late
orders['is_late'] = (orders['delivery_delay_days'] > 0).astype(int)
```

```python
orders['is_late']
```

```python
orders['order_id'].nunique()
```

```python
orders_sellers = orders.merge(
    order_items[['order_id', 'seller_id']],
    on='order_id',
    how='left'
)
orders_sellers['seller_id'].isnull().sum()
seller_delay = orders_sellers.groupby('seller_id').agg({
    'order_id': 'count',
    'is_late': 'sum'
}).reset_index()
seller_delay['delay_pct'] = (
    seller_delay['is_late'] / seller_delay['order_id']
) * 100

seller_delay = seller_delay[seller_delay['order_id'] >= 20]


seller_delay = seller_delay.sort_values(
    by='delay_pct',
    ascending=False
)

seller_delay.head(10)

top_sellers = seller_delay.head(10)

sns.barplot(
    x='delay_pct',
    y='seller_id',
    data=top_sellers
)

plt.title("Top 10 Sellers by Delay %")
plt.xlabel("Delay Percentage")
plt.ylabel("Seller ID")
plt.show()
```

```python
# Step 1: Merge orders + reviews
satisfaction = orders.merge(
    reviews[['order_id', 'review_score']],
    on='order_id',
    how='left'
)

# Step 2: Average rating for on-time vs late

on_time_rating = satisfaction[satisfaction['is_late'] == 0]['review_score'].mean()
late_rating = satisfaction[satisfaction['is_late'] == 1]['review_score'].mean()

print("On-time Rating:", on_time_rating)
print("Late Rating:", late_rating)
```

```python
sns.boxplot(x='is_late', y='review_score', data=satisfaction)
plt.title("Delay vs Customer Rating")
plt.show()
```
