```python
print(orders.shape)
print(orders.isnull().sum())
print(orders.duplicated().sum())
```

```python
date_cols = [
    'order_purchase_timestamp',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col])

orders = orders[orders['order_delivered_customer_date'].notna()]
```

```python
orders['delivery_days'] = (
    orders['order_delivered_customer_date'] - 
    orders['order_purchase_timestamp']
).dt.days

orders['delivery_delay_days'] = (
    orders['order_delivered_customer_date'] - 
    orders['order_estimated_delivery_date']
).dt.days

orders['is_late'] = (orders['delivery_delay_days'] > 0).astype(int)
```

```python
Q1 = orders['delivery_days'].quantile(0.25)
Q3 = orders['delivery_days'].quantile(0.75)
IQR = Q3 - Q1

orders = orders[
    (orders['delivery_days'] >= Q1 - 1.5*IQR) &
    (orders['delivery_days'] <= Q3 + 1.5*IQR)
]
```

```python
payments_agg = payments.groupby('order_id')['payment_value'].sum().reset_index()
payments_agg 
```

```python
df = orders.merge(customers, on='customer_id', how='left')
df = df.merge(order_items, on='order_id', how='left')
df = df.merge(products, on='product_id', how='left')
df = df.merge(payments_agg, on='order_id', how='left')
df = df.merge(reviews, on='order_id', how='left')
```

```python
print(df.shape)
print(df.isnull().sum().sort_values(ascending=False).head(10))
print("Duplicates:", df.duplicated().sum())
```

```python
print(df.shape)
print(df.isnull().sum().sort_values(ascending=False).head(10))
print("Duplicates:", df.duplicated().sum())
```

```python
df['review_score'] = df['review_score'].fillna(df['review_score'].median())
df['payment_value'] = df['payment_value'].fillna(0)
```

```python
kpis = {
    "Total Orders": df['order_id'].nunique(),
    "Total Revenue": df['payment_value'].sum(),
    "Avg Order Value": df['payment_value'].mean(),
    "Late Delivery %": df['is_late'].mean()*100
}

print(kpis)
```
